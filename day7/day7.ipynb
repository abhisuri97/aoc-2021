```python
def read_input(file):
    pos = []
    with open(file) as f:
        for line in f:
            pos = [int(x) for x in line.rstrip().split(',')]
    return pos
read_input('input.txt')
```

```python
def day7_pt1(file):
    pos = read_input(file)
    min_fuel = float('inf')
    for i in range(max(pos) + 1):
        fuel = 0
        for a in pos:
            fuel += abs(a-i)
        if fuel < min_fuel:
            min_fuel = fuel
            
    return min_fuel
day7_pt1('input.txt')    
```

```python
def get_triangular(num):
    return (num * (num + 1))/2

def day7_pt2(file):
    pos = read_input(file)
    min_fuel = float('inf')
    for i in range(max(pos) + 1):
        fuel = 0
        for a in pos:
            fuel += get_triangular(abs(a-i))
        if fuel < min_fuel:
            min_fuel = fuel
            
    return min_fuel

day7_pt2('input.txt')
```
