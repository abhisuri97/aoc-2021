```python
def read_input(file):
    arr = []
    with open(file) as f:
        for line in f:
            arr.append(list(line.rstrip()))
    return arr

# read_input('input.txt')
```

```python
def day10_pt1(file):
    arr = read_input(file)
    va_arr = [3, 57, 1197, 25137] 
    er_arr = [0, 0, 0, 0] # ), ], }, >
    for a in arr:
        symbols = []
        for v in a:
            if v in '({<[':
                symbols.append(v)
                continue
            else:
                if v == ')':
                    if symbols[-1] == '(':
                        symbols.pop()
                    else:
                        er_arr[0] += 1
                        break
                if v == ']':
                    if symbols[-1] == '[':
                        symbols.pop()
                    else:
                        er_arr[1] += 1
                        break
                if v == '}':
                    if symbols[-1] == '{':
                        symbols.pop()
                    else:
                        er_arr[2] += 1
                        break
                if v == '>':
                    if symbols[-1] == '<':
                        symbols.pop()
                    else:
                        er_arr[3] += 1
                        break
            
    s = 0
    for idx, e in enumerate(er_arr):
        s += e * va_arr[idx]
    return s
    
day10_pt1('input.txt')
```

```python
def day10_pt2(file):
    arr = read_input(file)
    va_arr = [3, 57, 1197, 25137] 
    er_arr = [0, 0, 0, 0] # ), ], }, >
    bad_lines = []
    for idx,a in enumerate(arr):
        symbols = []
        for v in a:
            if v in '({<[':
                symbols.append(v)
                continue
            else:
                if v == ')':
                    if symbols[-1] == '(':
                        symbols.pop()
                    else:
                        er_arr[0] += 1
                        bad_lines.append(idx)
                        break
                if v == ']':
                    if symbols[-1] == '[':
                        symbols.pop()
                    else:
                        er_arr[1] += 1
                        bad_lines.append(idx)
                        break
                if v == '}':
                    if symbols[-1] == '{':
                        symbols.pop()
                    else:
                        er_arr[2] += 1
                        bad_lines.append(idx)
                        break
                if v == '>':
                    if symbols[-1] == '<':
                        symbols.pop()
                    else:
                        er_arr[3] += 1
                        bad_lines.append(idx)
                        break
    scores = []
    for idx, a in enumerate(arr):
        score = 0
        if idx not in bad_lines:
            symbols = []
            for v in a:
                if v in '({<[':
                    symbols.append(v)
                    continue
                if v in ')}>]':
                    symbols.pop()
                    continue
            while len(symbols) > 0:
                v = symbols.pop()
                
                add = 0
                if v == '(':
                    add = 1
                if v == '[':
                    add = 2
                if v == '{':
                    add = 3
                if v == '<':
                    add = 4
                score = (score * 5) + add
        scores.append(score)
    return sorted(scores)[int(len(scores)/2)]
    
day10_pt2('inputalt.txt')
```
